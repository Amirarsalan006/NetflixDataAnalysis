```python
# Read the Data

df= pd.read_csv(r'D:\Projects\Python\Netflix\movies.csv')
pd.set_option('display.max_rows', None)
df.head()
```

```python
#checking missing values

for col in df.columns:
    miss_value = (df[col].isna().sum()/df['name'].count()) * 100
    print ('{} : {:.2f}%'.format(col,miss_value))

```

```python
# Updating missing values


#For Rating column

df['rating'].drop_duplicates()
```

```python
df['rating'] = df['rating'].fillna('Not Rated')
df['rating'] = df['rating'].replace({'Approved':'Not Rated','Unrated':'Not Rated'})
```

```python
# For Score Column

print(df['score'].isna().sum())
print(df['votes'].isna().sum())
```

```python
df['score'] = df['score'].fillna(0)
df['votes'] = df['votes'].fillna(0)
```

```python
df['writer'] =df['writer'].fillna('Unknown')
df['star'] =df['star'].fillna('Unknown')
df['country'] =df['country'].fillna('Unknown')
```

```python
#Checking missing values again for further actions

for col in df.columns:
    miss_value = (df[col].isna().sum()/df['name'].count()) * 100
    print('{} {:.2f}%'.format(col,miss_value))
```

```python
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
df['runtime'] = df['runtime'].fillna(0)
```

```python
df['genre'] = df['genre'].fillna('Unknown')
df['company'] = df['company'].fillna('Unknown')

df['genre'] = df['genre'].astype('category')
df['company'] = df['company'].astype('category')

```

```python
df['released_year'] = df['released'].str.extract(r'\b(\d{4})\b')
df['released_year'] = df['released_year'].fillna(0000)
df.head(10)
```

```python
# check missing values for the last time

for col in df.columns:
    miss_value = (df[col].isna().sum()/df['name'].count()) * 100
    print('{} {:.2f}%'.format(col,miss_value))
```
